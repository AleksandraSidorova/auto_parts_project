```python
import pandas as pd
```

```python
df1 = pd.read_csv('Июнь_2024_1.csv', sep=';')
```

```python
df2 = pd.read_csv('Июнь_2024_2.csv', sep=';')
```

```python
df3 = pd.read_csv('Июнь_2024_3.csv', sep=';')
```

```python
print(df1)
```

```python
df_joined = pd.concat([df1, df2, df3], axis=0) #axis=0 удаляет первые СТРОКИ (заголовки) в дата-фреймах
df_joined
```

```python
df_joined.reset_index(drop=True) #обнуляет индексы каждого дата-фрейма и делает по порядку
```

```python
df_joined.columns
```

```python
df_work = df_joined.drop(df_joined.columns[[3, 4, 5, 6, 7]], axis = 1) #axis=1 удаляет СТОЛБЦЫ по номерам с 0-го
df_work
```

```python
pd.set_option('display.max_columns', None) #показывает все столбцы
#pd.reset_option('display.max_columns') #отменить вывод всех столбцов
```

```python
df_work.drop(df_work.columns[[38, 39]], axis = 1, inplace = True) #inplace=true удаление с заменой в дата-фрейме
df_work
```

```python
df_work.columns
```

```python
df_work.drop(columns = ['Куратор клиента', 'Направление', 'Город', 'Куратор поставщик товара', 'Процент скидки', 'Филиал', 'Куратор Торгового Отдела', 'Категория 2', 'Категория 3', 'Код бренда', 'Код поставщика', 'Код группы', 'Номер Реализации', 'Код склада', 'Вес'], axis = 1, inplace = True) #удаление по названию столбца
df_work
```

```python
print(df_work)
```

```python
df_work.to_csv('Июнь_2024_joined.csv', index=False)
```
